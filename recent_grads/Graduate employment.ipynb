# Unemployment analysis

Using FiveThirtyEight's cleaned dataset for graduate employment rates per subject. Documentation for the dataset can be found <a href="https://github.com/fivethirtyeight/data/tree/master/college-majors">on their Github</a>.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')

recent_grads.rename({
    'Rank': 'rank',
    'Major_code': 'major_code',
    'Major': 'major',
    'Total': 'total',
    'Men': 'men',
    'Women': 'women',
    'Major_category': 'major_category',
    'ShareWomen': 'share_women',
    'Sample_size': 'sample_size',
    'Employed': 'employed',
    'Full_time': 'full_time',
    'Part_time': 'part_time',
    'Full_time_year_round': 'full_time_year_round',
    'Unemployed': 'unemployed',
    'Unemployment_rate': 'unemployment_rate',
    'Median': 'median',
    'P25th': 'p25th',
    'P75th': 'p75th',
    'College_jobs': 'college_jobs',
    'Non_college_jobs': 'non_college_jobs',
    'Low_wage_jobs': 'low_wage_jobs',
}, axis='columns', inplace=True)

recent_grads.iloc[0]
```
